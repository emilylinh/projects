```python
def rk4(N):
    w = alpha 
    t = a 
    sol = [w] 
    for i in range(N-1):
        k1 = h*f(t, w)
        k2 = h*f(t + h/2, w + k1/2)
        k3 = h*f(t + h/2, w + k2/2)
        k4 = h*f(t + h, w + k3)
        w = w + 1/6*(k1 + 2*k2 + 2*k3 + k4)
        t += h 
        sol.append(w)
    return sol 
```
