2\.  Use the Modified Euler method to approximate the solutions to each of the following initial-value problems and compare the results to the actual values.

```python
from math import*
import pandas as pd 

def actual(a, h, N):
    t = a 
    solutions = []
    for i in range(N+1):
        solutions.append(y(t))
        t = t + h      
    return solutions 

def modifiedEuler(a, b, N, alpha):
    h = (b-a)/N
    t = a
    w = alpha 
    solutions = [w]
    for i in range(N):
        t1 = t + h
        w = w + h/2*(f(t, w) + f(t1, w+h*f(t,w)))        
        t += h
        solutions.append(w)
    return solutions

def t_i(a, h, N):
    t = a
    lst = [a]
    for i in range(N):
        t = t + h 
        lst.append(t)
    return lst 
```

$\;\;\;\;$ b. $y' = \frac{1+t}{1+y}, \; 1\leq t \leq 2, \; y(1)=2,$ with $h=0.5$; actual solution $y(t) = \sqrt{t^2+2t+6}-1$.

```python
y = lambda t: (t**2 + 2*t + 6)**(1/2) - 1        

def f(t, y):
    return (1+t)/(1+y)


data = {'t_i':t_i(1, 0.5, 2),
        'w_i (approx. values)':modifiedEuler(1, 2, 2, 2), 
       'y[t_i] (exact values)':actual(1, 0.5, 2)}
table = pd.DataFrame(data)
table['actual error'] = abs(table['y[t_i] (exact values)'] - table['w_i (approx. values)'])
table
```

11\. Repeat Exercise 3 using Heun's method.

```python
def Huen(a, b, N, alpha):
    h = (b-a)/N
    t = a
    w = alpha 
    solutions = [w]
    for i in range(N):
        w = w + h/4*(f(t, w) + 3*(f(t+2*h/3, w+2*h/3*f(t+h/3, w+h/3*f(t, w)))))    
        t += h
        solutions.append(w)
    return solutions
```

15\. Repeat Exercise 3 using the Runge-Kutta method of order four.

```python
def runge_kutta_o4(a, b, N, alpha):
    h = (b-a)/N
    t = a
    w = alpha 
    solutions = [w]
    for i in range(N):
        t1 = t + h
        k1 = h*f(t, w)
        k2 = h*f(t + h/2, w + 1/2*k1)
        k3 = h*f(t + h/2, w + 1/2*k2)
        k4 = h*f(t1, w + k3)
        w = w + 1/6*(k1 + 2*k2 + 2*k3 + k4)
        t += h
        solutions.append(w)
    return solutions
```

The Runge-Kutta method of order four can be written in the form $$\begin{eqnarray} w_0 &=& \alpha, \\ w_{i+1} &=& w_i +\frac{h}{6}f(t_i, w_i) + \frac{h}{3}f(t_1+\alpha_1h, \; w_i+\delta_1hf(t_i,w_i))\\ &+& \frac{h}{3}f(t_i+\alpha_2h,\; w_i+\delta_2hf(t_i+\gamma_2h,\; w_i+\gamma_3hf(t_i, w_i))) \\ &+& \frac{h}{6}f(t_i+\alpha_3h,\; w_i+\delta_3hf(t_i+\gamma_4h,\; w_i+\gamma_5hf(t_i+\gamma_6h,\; w_i+\gamma_7h(t_i,w_i)))). \end{eqnarray}$$ Find the values of the constants $$\alpha_1,\alpha_2,\alpha_3,\delta_1,\delta_2,\delta_3,\gamma_2,\gamma_3,\gamma_4,\gamma_5,\gamma_6,\; \text{and } \gamma_7.$$

$\;\;\;\;$_Sol._ Given that $$w_0=\alpha \\ k_1 = h\dot f(t_i, w_i) \\ k_2 = h\dot f(t_i+\frac{h}{2}, w_i+\frac{1}{2}k_1) \\ k_3 = h\dot f(t_i+\frac{h}{2}, w_i+\frac{1}{2}k_2) \\ k_4 = h\dot f(t_i+h, w_i+k_3) \\ w_{i+1} = w_i + \frac{1}{6}[k_1+2k_2+2k_3+k_4]$$ for $i=0,1..., N-1$, we could simply the result to $$\begin{eqnarray} w_{i+1}&=& w_i+\Bigg[\frac{h\dot f(t_i, w_i)}{6} + \frac{h\dot f(t_i+\frac{h}{2}, w_i+\frac{1}{2}k_1)}{3} \\  &+& \frac{h\dot f(t_i+\frac{h}{2}, w_i+\frac{1}{2}k_2)}{3} + \frac{h\dot f(t_i+h, w_i+k_3)}{6}\Bigg].\end{eqnarray}$$

And furthermore, $$\begin{eqnarray} w_{i+1}&=&w_i+\Bigg[\frac{h\dot f(t_i, w_i)}{6}+\frac{h\dot f(t_i+\frac{h}{2}, w_i+\frac{1}{2}h\dot f(t_i, w_i))}{3} \\ &+& \frac{h\dot f(t_i+\frac{h}{2}, w_i+\frac{1}{2}h\dot f(t_i+\frac{h}{2}, w_i+\frac{1}{2}k_1))}{3} \\ &+& \frac{h\dot f(t_i+h, w_i+ h\dot f(t_i+\frac{h}{2}, w_i+\frac{1}{2}k_2))}{6}\Bigg] \\ &=&w_i+\frac{h}{6}\dot f(t_i, w_i)+\frac{h}{3}\dot f\Bigg(t_i+\frac{h}{2}, w_i+\frac{1}{2}h\dot f(t_i, w_i) \Bigg) \\ &+& \frac{h}{3}\dot f\Bigg(t_i+\frac{h}{2}, w_i+\frac{1}{2}h\dot f\Big(t_i+\frac{h}{2}, w_i+\frac{1}{2}h\dot f(t_i, w_i)\Big)\Bigg) \\ &+& \frac{h}{6}\dot f\Bigg(t_i+h, w_i+ h\dot f\Big(t_i+\frac{h}{2}, w_i+\frac{1}{2}h\dot f\big(t_i+\frac{h}{2}, w_i+\frac{1}{2}h\dot f(t_i, w_i)\big)\Big)\Bigg). \end{eqnarray}$$

Comparing the form above to the given results: $$\begin{eqnarray}w_{i+1} &=& w_i +\frac{h}{6}f(t_i, w_i) + \frac{h}{3}f(t_1+\alpha_1h, \; w_i+\delta_1hf(t_i,w_i))\\ &+& \frac{h}{3}f(t_i+\alpha_2h,\; w_i+\delta_2hf(t_i+\gamma_2h,\; w_i+\gamma_3hf(t_i, w_i))) \\ &+& \frac{h}{6}f(t_i+\alpha_3h,\; w_i+\delta_3hf(t_i+\gamma_4h,\; w_i+\gamma_5hf(t_i+\gamma_6h,\; w_i+\gamma_7h(t_i,w_i)))). \end{eqnarray}$$ 

We find that $$\alpha_1 = \frac{1}{2},\alpha_2 = \frac{1}{2},\alpha_3=1,\\ \delta_1=\frac{1}{2},\delta_2=\frac{1}{2},\delta_3=1, \\ \gamma_2=\frac{1}{2},\gamma_3=\frac{1}{2},\gamma_4=\frac{1}{2}, \\ \gamma_5=\frac{1}{2},\gamma_6=\frac{1}{2},\; \text{and } \gamma_7=\frac{1}{2}.$$
