## Characterizing the Rate of Returns using Gaussian Quadrature and Riskiness of Assets/Portfolios

To start off, we first define the characteristics of assets/portfolios. The first common characterization is through the average rate of return of an asset/portfolio over a period of time which is defined as $\mu = E(\rho)$. Using the average rate of return, we could capture the "expected" rate of return; however, it doesn't indicate the certainity or how large the actual return may deviate from this expected rate of return. Thus, to do so and to capture the riskiness of the portfolio, we would use the variance, $\sigma^2 = \text{Var}(\rho) = E(|\rho - \mu|^2)$.

Given the statistics of each assets above, we will now go on to define the characteristics of our portfolio. We define $W = \sum_{i = 1}^n W_i$ to be the total amount of wealth used to invest in $n$ assets where $W_i$ denotes the amount of money invested in $i^{th}$ asset for $i \in 1,..., n$. For the the $i^{th}$ asset, let $w_i = \frac{W_i}{W}$ be the fraction of wealth invested and $S_i(t)$ be the index value at time $t$. Then, the value of the portfolio at time $t$ could be denoted as $Q_p(t) = \sum_{i=1}^n \frac{W_i}{S_i(0)}S_i(t),$ where $Q_p(0) = W$. Using the rate of change formula, we could define the rate of return of the portfolio $\rho_p$ at time $t$ to be $$\begin{eqnarray} \rho_p(t) &=& \frac{Q_p(t) - Q_p(0)}{Q_p(0)} \\ &=& \frac{\sum_{i=1}^n \frac{W_i}{S_i(0)}S_i(t) - W}{W} \\ &=& \sum_{i=1}^n \frac{W_i}{WS_i(0)}S_i(t) - \sum_{i=1}^n \frac{W_i}{W} \\ &=& \sum_{i=1}^n w_i \frac{(S_i(t) - S_i(0))}{S_i(0)} \\ &=& \sum_{i=1}^n w_i\rho_i. \end{eqnarray}$$ Thus, the rate of return of portfolio is the weighted average of the rates of returns of the assets.  


### Defining Assets
Let $R_{i}$ (i=A,B,C) denote the return on asset i   

$w_{i}$ denote the share of wealth invested in asset i, or weights  

Thus, we will assume that all wealth is invested into only these three assets  

$R_{p,w}= w_{a}R_{a} + w_{b}R_{b}+w_{c}R_{c}$ where $R_{p,w}$ is the random variable for portfolio return  

The expected return on the portfolio is 
$\mu_{p,w} = \mathbb{E}[R_{p,w}] = w_{a}\mu_{a} + w_{b}\mu_{b}+w_{c}\mu_{c}$  

The variance of the portfolio return is
$\sigma^2_{p,w} = var(R_{p,w})$  
= $w^2_{a}\sigma^2_{a} + w^2_{b}\sigma^2_{b}+w^2_{c}\sigma^2_{c}+ 2w_{a}w_{b}\sigma_{ab} + 2w_{a}w_{c}\sigma_{ac}+2w_{b}w_{c}\sigma_{bc}$  where $\sigma_{xy} = Cov(x,y)$ for $x, y\in a, b, c$.

Note: There will be 3 variance terms and 6 covariance terms.


To demonstrate these derivations, we used the following code below. Hypothetically for our parameters, let's say we bought each asset of our portfolio at its closing price on November 30$^{th}$, 1993 and would like to find the rate of return, expected return, and variance of our portfolio twenty-six years from now.  

```python
from math import *
# Porfolio Monthly Return Dataframe 
W = mfst['Adj Close'][0] + sbux['Adj Close'][0] + ba['Adj Close'][0]
w1 =  mfst['Adj Close'][0]/W
w2 = sbux['Adj Close'][0]/W
w3 = ba['Adj Close'][0]/W

mfstR = mfst['Monthly Return']
sbuxR = sbux['Monthly Return']
baR = ba['Monthly Return']
monthly_ret = pd.concat([mfstR, sbuxR, baR], axis = 1)
monthly_ret.columns = ['MFST', 'SBUX', 'BA']
monthly_ret['Portfolio'] = w1*mfstR + w2*sbuxR + w3*baR
print('The rate of returns for our portfolio is\n', monthly_ret['Portfolio'])

print('\nThe expected return for our portfolio is', sumEqn([w1,w2,w3], s['Expected Ret.'].tolist()))

portf_var1 = sumEqn([w1**2,w2**2,w3**2], s['Var.'].tolist())
portf_var2 = 2*(w1*w2*s['Cov.'][1] + w1*w3*s['Cov.'][0] + w2*w3*s['Cov.'][2])
print('\nThe variance of returns for our portfolio is', portf_var1+portf_var2)
```
